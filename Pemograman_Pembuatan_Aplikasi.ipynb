```python
import json

# Data produk makeup + stok
produk = {
    1: {"nama": "Lipstik Matte", "harga": 75000, "stok": 60},
    2: {"nama": "Foundation Cair", "harga": 120000, "stok": 15},
    3: {"nama": "Eyeliner Hitam", "harga": 45000, "stok": 30},
    4: {"nama": "Bedak Padat", "harga": 65000, "stok": 55},
    5: {"nama": "Blush On", "harga": 50000, "stok": 45}
}

keranjang = []

def tampilkan_produk():
    print("\n=== Daftar Produk Makeup ===")
    for kode, item in produk.items():
        print(f"{kode}. {item['nama']} - Rp{item['harga']:,} | Stok: {item['stok']}")

def tampilkan_keranjang():
    print("\n=== Keranjang Belanja ===")
    total = 0
    for item in keranjang:
        kode = item["kode"]
        jumlah = item["jumlah"]
        nama = produk[kode]["nama"]
        harga = produk[kode]["harga"]
        subtotal = harga * jumlah
        print(f"{nama} x{jumlah} = Rp{subtotal:,}")
        total += subtotal
    print(f"\nTotal Belanja: Rp{total:,}")
    return total

def simpan_struk_json(keranjang, total, uang_bayar, kembalian):
    isi_keranjang = []
    for item in keranjang:
        kode = item["kode"]
        nama = produk[kode]["nama"]
        harga = produk[kode]["harga"]
        jumlah = item["jumlah"]
        subtotal = harga * jumlah
        isi_keranjang.append({
            "nama": nama,
            "jumlah": jumlah,
            "subtotal": subtotal
        })

    struk = {
        "keranjang": isi_keranjang,
        "total": total,
        "uang_bayar": uang_bayar,
        "kembalian": kembalian
    }

    with open("struk_belanja.json", "w") as f:
        json.dump(struk, f, indent=4)
    print("‚úÖ Struk disimpan sebagai 'struk_belanja.json'")

def main():
    print("Selamat Datang di Toko Makeup GlowStore üíÑ")
    tampilkan_produk()

    while True:
        try:
            kode = int(input("\nMasukkan kode produk (0 untuk selesai): "))
            if kode == 0:
                break
            if kode not in produk:
                print("‚ùå Kode produk tidak tersedia.")
                continue
            jumlah = int(input("Jumlah: "))
            if jumlah <= 0:
                print("‚ùå Jumlah harus lebih dari 0.")
                continue
            if jumlah > produk[kode]["stok"]:
                print("‚ùå Stok tidak mencukupi.")
                continue
            keranjang.append({"kode": kode, "jumlah": jumlah})
            produk[kode]["stok"] -= jumlah  # kurangi stok
        except ValueError:
            print("‚ùå Masukkan angka yang valid.")

    if keranjang:
        total = tampilkan_keranjang()

        # Input uang bayar
        while True:
            try:
                bayar = int(input("Masukkan jumlah uang pembeli: Rp"))
                if bayar < total:
                    print("‚ùå Uang tidak cukup. Total belanja Rp{:,.0f}".format(total))
                else:
                    break
            except ValueError:
                print("‚ùå Masukkan nominal angka saja.")

        kembalian = bayar - total
        print(f"‚úÖ Kembalian: Rp{kembalian:,}")

        simpan_struk_json(keranjang, total, bayar, kembalian)
    else:
        print("Keranjang kosong. Terima kasih telah berkunjung!")

if __name__ == "__main__":
    main()
```
